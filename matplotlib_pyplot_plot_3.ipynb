```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)

y1 = x * 1.5
y2 = x * 2
y3 = x * 2.5
y4 = x * 3
y5 = x * 3.5
```

```python
fig = plt.figure(dpi=400)

plt.plot(x, y1, 'y--', label='y1')
plt.plot(x, y2, 'b*', label='y2')
plt.plot(x, y3, 'g:', label='y3')
plt.plot(x, y4, 'r.-', label='y4')
plt.plot(x, y5, 'm-', label='y5')

plt.legend()
plt.show()
plt.close()
```

```python
fig = plt.figure(dpi=400)

plt.plot(x, y1, 'y--',
         x, y2, 'b*',
         x, y3, 'g:',
         x, y4, 'r.-',
         x, y5, 'm-')

plt.show()
plt.close()
```
