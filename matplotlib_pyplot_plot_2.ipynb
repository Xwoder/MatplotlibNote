```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(dpi=400)
plt.plot(
    [0, 2, 4, 6, 8],  # x 轴数据
    [3, 1, 4, 5, 2]  # y 轴数据
)
plt.ylabel('Grade')

plt.axis([-1,  # x轴起始坐标值
          9,  # x轴终止坐标值
          0,  # y轴起始坐标值
          6,  # y轴终止坐标值
          ])

plt.show()
plt.close()
```
